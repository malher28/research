```python
import pandas as pd
import numpy as np
```

```python
# set seed
np.random.seed(42)

# number of observations
n = 10000

# number of predictors
p = 10

# make array of dataset
X = np.random.normal(size=(n, p))

# generate target -> 5% is 1's
target = np.random.binomial(1, 0.05, size=n)

# make categorical predictor -> planet
planet = np.random.uniform(0, 1, n)

# choose distribution of planet
planet = np.select(
    [
        planet < 0.3,
        (planet >= 0.3) & (planet < 0.9),
        planet >= 0.9
    ],
    [0, 1, 2])

X[:, 0] = planet

# change up distributions of specific columns
X[:, [2, 3, 8]] = np.random.uniform(0, 1, size=(n, 3))
X[:, 1] = np.random.randint(18, 25, n)
X[:, 4] = np.random.normal(0, 1, n)
X[:, 7] = np.random.exponential(1, n)



# convert to dataframe
df = pd.DataFrame(X, columns=[f"X{i+1}" for i in range(p)])
df["target"] = target

# name predictors
df = df.rename(columns={
    "X1": "planet",   
    "X2": "age"
})

df.head()
```

```python
# individual counts
print(df["planet"].value_counts())
print(df['target'].value_counts())
```

```python
# percentage of planet and target
pd.crosstab(df["planet"], df["target"], normalize='index')
```

```python
# save dataframe for use
df.to_csv("toy_data.csv", index=False)
```
