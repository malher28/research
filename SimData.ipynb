```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from scipy.sparse import csr_matrix, save_npz

# set seed
np.random.seed(42)

# number of observations
n = 10000

# number of predictors
p = 10

# make array of dataset
X = np.random.normal(size=(n, p))

# make categorical predictor -> planet
planet_raw = np.random.uniform(0, 1, n)

# planet distribution
planet = np.select(
    [
        planet_raw < 0.30,
        (planet_raw >= 0.30) & (planet_raw < 0.90),
        planet_raw >= 0.90
    ],
    [0, 1, 2]
)

X[:, 0] = planet

# probability of target = 1 depends on planet
target_probability = np.select(
    [
        planet == 0,
        planet == 1,
        planet == 2
    ],
    [
        0.05,  # .5 ones for planet 0
        0.05,  # .5 ones for planet 1
        0.01   # .1 ones for planet 2
    ]
)

target = np.random.binomial(
    1,
    target_probability,
    size=n
)

# change up distributions of specific columns
X[:, [2, 3, 8]] = np.random.uniform(0, 1, size=(n, 3))
X[:, 1] = np.random.randint(18, 25, n)
X[:, 4] = np.random.normal(0, 1, n)
X[:, 7] = np.random.exponential(1, n)



# convert to dataframe
df = pd.DataFrame(X, columns=[f"X{i+1}" for i in range(p)])
df["target"] = target

# name predictors
df = df.rename(columns={
    "X1": "planet",   
    "X2": "age"
})

# convert to categories
df["planet"] = df["planet"].map({
    0: "A",
    1: "B",
    2: "C"
})

# df.head()

# individual counts
print(df["planet"].value_counts())
print(df['target'].value_counts())

# percentage of planet and target
print(pd.crosstab(df["planet"], df["target"], normalize='index'))

```

```python
df.to_csv("simdata.csv", index=False)
```
